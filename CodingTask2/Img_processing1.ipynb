```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Read the image
wellplate = cv2.imread("image_png\IMG_5379.png")

# Crop the image to the middle half
wellplate = wellplate[wellplate.shape[0] // 2:, :]

# Apply the Gaussian filter
wellplate_f = scipy.ndimage.gaussian_filter(wellplate, sigma=5)

# Convert the image to float32 if necessary
wellplate_f = wellplate_f.astype(np.float32)

# Apply the Sobel filter in the X direction
sobel_x = cv2.Sobel(wellplate_f, cv2.CV_64F, 1, 0, ksize=5)

# Apply the Sobel filter in the Y direction
sobel_y = cv2.Sobel(wellplate_f, cv2.CV_64F, 0, 1, ksize=5)

# Normalize the Sobel outputs to the range [0, 255] for display
sobel_x_normalized = ((sobel_x - sobel_x.min()) / (sobel_x.max() - sobel_x.min()) * 255).astype(np.uint8)
sobel_y_normalized = ((sobel_y - sobel_y.min()) / (sobel_y.max() - sobel_y.min()) * 255).astype(np.uint8)
sobel_combined = cv2.addWeighted(sobel_x, 0.2, sobel_y, 0.9, 0)


thresh = cv2.threshold(sobel_combined,40, 255, cv2.THRESH_BINARY_INV)[1]
# Display the images using subplots
thresh = thresh[:,:,1].astype(np.uint8)
plt.imshow(thresh)
print(type(thresh))
```

```python
import os
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

h, w = thresh.shape[:2]
imFlood = thresh.copy()
mask = np.zeros((h+2, w+2), np.uint8)
new_img = cv2.floodFill(imFlood, mask, (0,0), 0)

imFlood[np.where(thresh==0)]=255
imFlood=~imFlood

cnts, _ = cv2.findContours(imFlood, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
for cnt in cnts:
    peri = cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, 0.02*peri, True)
    x, y, w, h = cv2.boundingRect(approx)
    bob = wellplate[y:y+h, x:x+w]
    cv2.rectangle(wellplate, (x, y), (x+w, y+h), (0, 255, 0), 2)

cv2.imwrite("im_6_output.png",wellplate)

plt.imshow(thresh)
```

```python
a = 5
wellplate_b = cv2.GaussianBlur(wellplate,(a,a),10)
edges = cv2.Canny(wellplate_b,20,60)
laplacian = cv2.Laplacian(wellplate_b,cv2.CV_64F)
l_t = cv2.threshold(laplacian,2, 255, cv2.THRESH_BINARY_INV)[1]
l_t = l_t[:,:,1].astype(np.uint8)
edges = cv2.Canny(l_t,100,200)




plt.imshow(edges)
```
